### Identify and handle outliers

### Detact Outlier

```python

#Method to handle outliers: IQR,Boxplot
import seaborn as sns
import matplotlib.pyplot as plt

numeric_value=['age','height','weight','ap_hi','ap_lo','cholesterol','gluc','smoke','alco','active','cardio']


for col in numeric_value:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower_limit) | (df[col] > upper_limit)]
    print(f"{col} → Outliers found: {outliers.shape[0]}")
```

### Remove Outlier

```python
for col in numeric_value:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
df = df[(df[col] >= lower_limit) & (df[col] <= upper_limit)]
```

###  Encode categorical variables Normalize/scale numerical features

```python
#convert data in in range 0 to 1
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[numeric_value] = scaler.fit_transform(df[numeric_value])
df[numeric_value]
```

```python
#convert data to mean=0 and std=1
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[numeric_value] = scaler.fit_transform(df[numeric_value])
df[numeric_value]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
```

### Display Graph

```python
import seaborn as sns
import matplotlib.pyplot as plt

features = ['age','height','weight','ap_hi','ap_lo','cholesterol','gluc']

for col in features:
    plt.figure(figsize=(8,4))
    sns.histplot(data=df, x=col, hue='cardio', kde=True, bins=30)
    #kde use for smooth curve
    plt.title(f"Histogram of {col} vs Heart Disease")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```
