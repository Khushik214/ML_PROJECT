# Step 8. A boxplot can also be used to show the distribution of values for each attribute.

```python
import matplotlib.pyplot as plt

data[['height','weight']].boxplot()
plt.show()
```

# Step 9. How is the dataset indexed?

# Step 10.  Display scatter plot for any 5 pair of attributes , we can use a scatter plot to visualize their joint distribution

```python
import matplotlib.pyplot as plt
x=data['height']
y=data['weight']
plt.scatter(x,y)
plt.show()
```

# Step 11. which age have been weight more than 100?

```python
data=data.groupby('age')['weight'].sum().loc[lambda x: x > 100]
data
```

# Step 12. Accers rows by their integer locations


```python
data.iloc[0]
```

# Step 13. convert age days in to year

```python
data['age']=data['age']//365
data
```

# Step 14 Display the histogram for Age attribute by discretizing it into 8 separate bins and counting the frequency for each bin.

```python
import matplotlib.pyplot as plt

data['age'].hist(bins = 10, edgecolor = 'green')
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('histogram')
plt.show()
```
